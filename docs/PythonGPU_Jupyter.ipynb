The following is how to use one of the ML scicomp machines that
has 4 Titan RTX GPU cards installed. A couple of notes:

1. Setting up the software environment seems to be more easily
done using conda. Will Phelps sent me the magic commands.

1. It seems at least part of the environment is maintained
when the srun command is called. At least I was still in the
same working directory and python3 was pointing to the venv.
However, the virtual environment was not completely set up
so I had to source the activate script again

2. I was able to pip install from the sciml190X node whereas
it failed to do it from the regular ifarm node due to missing
SSL modules.


----------------------------------------------------------

Setting up:
-------------------------
This part only needs to be done once to get the software.

The software must be set up using a computer other than sciml190X
since it needs a level of outside network access not available
there. Also, the size of the installation is large enough that
it won't fit easily in you home directory. Conda likes to install
things in ~/.conda so that must be a link to some larger disk.

ssh ifarm1802
ln -s /work/halld2/home/davidl/builds/CONDA_ENV ~/.conda

bash
source /etc/profile.d/modules.sh
module use /apps/modulefiles
module load anaconda2/4.5.12
conda create -n tf-gpu tensorflow-gpu=1.14 cudatoolkit=10.0 keras numpy=1.16.4
conda activate tf-gpu
pip install Pillow
conda install ipython matplotlib



Logging in to sciml190X
-------------------------

To login to one of the sciml190X nodes do the following:

salloc --gres gpu:TitanRTX:2 --partition gpu --nodes 1
srun --pty bash


source /etc/profile.d/modules.sh
module use /apps/modulefiles
module load anaconda2/4.5.12
conda activate tf-gpu

If you log out, of both the srun and salloc command, then 
the job should complete and the resource should be released.
You can check this by just running the sacct command to see 
a list of your jobs and if there are any running:

sacct

If there is a running job that you want to kill so the resource is 
released, cancel the jobid via:

scancel <jobid>


To see available devices:

nvidia-smi


To see which devices were assigned to you so this:

printenv CUDA_VISIBLE_DEVICES
